```python
#Data Exploration 
#Loading Dataset


# Import the pandas library for data manipulation and analysis
import pandas as pd  

 # Import the requests library to send HTTP requests
import requests 

# Import BytesIO to handle byte streams
from io import BytesIO  

# Define the URL of the Excel file stored on Google Sheets
url = 'https://docs.google.com/spreadsheets/d/1KagwoQLy1quKvT_82amuS-x3UnsoIX4J6p02ewbjQNA/export?format=xlsx'

# Send a GET request to fetch the content of the URL
response = requests.get(url)

 # Check if the request was successful and raise an error if not
response.raise_for_status() 

# Use BytesIO to read the content of the response as a file-like object
excel_file = BytesIO(response.content)

# Load the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file)

# Display the first few rows of the DataFrame to verify the data was loaded correctly
df.head()
```

```python
# Data Visualization
# Importing the necessary libraries for data visualization

# Matplotlib is a plotting library for creating static, animated, and interactive visualizations in Python.
import matplotlib.pyplot as plt

# Seaborn is a data visualization library based on matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics.
import seaborn as sns
```

```python
# Creating Visualization 
# Sales Over Time


# Create a new figure for the plot with a specified size
plt.figure(figsize=(12, 6)) 

# Group the data by 'Order Date' and sum the 'Sales' for each date, then plot the result
df.groupby('Order Date')['Sales'].sum().plot()

# Set the title of the plot
plt.title('Total Sales Over Time')

# Label the x-axis
plt.xlabel('Date')

# Label the y-axis
plt.ylabel('Sales')

# Display the plot
plt.show()
```

```python
# Creating Visualization of Top Selling Products

# Create a new figure for the plot with a specified size
plt.figure(figsize=(12, 6))

# Group the data by 'Product Name' and sum the 'Sales' for each product,
# then select the top 10 products with the highest sales and plot the result as a bar chart
df.groupby('Product Name')['Sales'].sum().nlargest(10).plot(kind='bar')

# Set the title of the plot
plt.title('Top 10 Selling Products')

# Label the x-axis
plt.xlabel('Product')

# Label the y-axis
plt.ylabel('Sales')

# Display the plot
plt.show()
```

```python
# Creating Visualization of Sales by regions

# Create a new figure for the plot with a specified size
plt.figure(figsize=(12, 6))

# Create a bar plot using Seaborn to show sales by region
# 'x' specifies the column for the x-axis (Region)
# 'y' specifies the column for the y-axis (Sales)
# 'data' specifies the DataFrame containing the data
sns.barplot(x='Region', y='Sales', data=df)

# Set the title of the plot
plt.title('Sales by Region')

# Label the x-axis
plt.xlabel('Region')

# Label the y-axis
plt.ylabel('Sales')

# Rotate the x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
